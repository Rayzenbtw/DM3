### –ó–∞–≤–¥–∞–Ω–Ω—è 1: –∫—Ä–∏–ø—Ç–æ–≥—Ä–∞—Ñ—ñ—è (–ï–ª—å-–ì–∞–º–∞–ª—å)

‚úîÔ∏è *–Ø–∫—â–æ –≤–∞—à –≤–∞—Ä—ñ–∞–Ω—Ç –ø—Ä–æ –∞–ª–≥–æ—Ä–∏—Ç–º –ï–ª—å-–ì–∞–º–∞–ª—è, —Ç–æ –∑–Ω–∞–π–¥—ñ—Ç—å —Å–≤–æ—ó –ø–∞—Ä–∞–º–µ—Ç—Ä–∏ –∞–ª–≥–æ—Ä–∏—Ç–º—É —É —Ñ–∞–π–ª—ñ –∑ –≤–∞—Ä—ñ–∞–Ω—Ç–∞–º–∏ —ñ –≤–∫–∞–∂—ñ—Ç—å —Ç—É—Ç:

1Ô∏è‚É£ –°—Ç–≤–æ—Ä—ñ—Ç—å –≤—ñ–¥–∫—Ä–∏—Ç–∏–π –∫–ª—é—á –∞–ª–≥–æ—Ä–∏—Ç–º—É –ï–ª—å-–ì–∞–º–∞–ª—è. –î–ª—è —Ü—å–æ–≥–æ

1) –∑–Ω–∞–π–¥—ñ—Ç—å $g$ -- –ø–µ—Ä–≤—ñ—Å–Ω–∏–π –∫–æ—Ä—ñ–Ω—å –∑–∞ –º–æ–¥—É–ª–µ–º $p$ (–¥–æ–¥–∞—Ç–∫–æ–≤–∞ –∑–∞–¥–∞—á–∞ 9.9 –∑ –∞—Ä–∫—É—à—ñ–≤). –î–ª—è –∑–Ω–∞—Ö–æ–¥–∂–µ–Ω–Ω—è –ø–µ—Ä–≤—ñ—Å–Ω–æ–≥–æ –∫–æ—Ä–µ–Ω—è –≤–∞–º —Ç–∞–∫–æ–∂ –∑–Ω–∞–¥–æ–±–∏—Ç—å—Å—è —Ñ—É–Ω–∫—Ü—ñ—è –ø—ñ–¥–Ω–µ—Å–µ–Ω–Ω—è –¥–æ —Å—Ç–µ–ø–µ–Ω—è –∑–∞ –º–æ–¥—É–ª–µ–º n.

2) –æ–±—á–∏—Å–ª—ñ—Ç—å $y = g^x \ {\rm mod}\ p$.

–í—ñ–¥–∫—Ä–∏—Ç–∏–º –∫–ª—é—á–µ–º —î —Ç—Ä—ñ–π–∫–∞ $(y, g, p)$.

```python
p = 103
x = 47
m = 3
```

```python
#write your code here
# —à–≤–∏–¥–∫–µ –ø—ñ–¥–Ω–µ—Å–µ–Ω–Ω—è –¥–æ —Å—Ç–µ–ø–µ–Ω—è –∑–∞ –º–æ–¥—É–ª–µ–º
def modpow(base, exp, mod):
    result = 1
    base %= mod
    while exp > 0:
        if exp % 2 == 1:
            result = (result * base) % mod
        base = (base * base) % mod
        exp //= 2
    return result

# —Ä–æ–∑–∫–ª–∞–¥ —á–∏—Å–ª–∞ n –Ω–∞ –ø—Ä–æ—Å—Ç—ñ –º–Ω–æ–∂–Ω–∏–∫–∏
def factorize(n):
    factors = set()
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors.add(d)
            n //= d
        d += 1
    if n > 1:
        factors.add(n)
    return factors

# –ø–æ—à—É–∫ –ø–µ—Ä–≤—ñ—Å–Ω–æ–≥–æ –∫–æ—Ä–µ–Ω—è modulo p
def primitive_root(p):
    phi = p - 1
    factors = factorize(phi)

    for g in range(2, p):
        ok = True
        for q in factors:
            # g^(phi/q) mod p –Ω–µ –ø–æ–≤–∏–Ω–Ω–æ –±—É—Ç–∏ 1 
            if modpow(g, phi // q, p) == 1:
                ok = False
                break
        if ok:
            return g
    return None


p = 103
x = 47
m = 3

#–æ–±—á–∏—Å–ª–µ–Ω–Ω—è
g = primitive_root(p)
y = modpow(g, x, p)

print("–ü–µ—Ä–≤—ñ—Å–Ω–∏–π –∫–æ—Ä—ñ–Ω—å g =", g)
print("y = g^x mod p =", y)
print("–í—ñ–¥–∫—Ä–∏—Ç–∏–π –∫–ª—é—á (y, g, p) =", (y, g, p))
```

2) –æ–±—á–∏—Å–ª—ñ—Ç—å $a = g^k\ {\rm mod}\ p$, $b = y^k\cdot m\ {\rm mod}\ p$

–ü–∞—Ä–∞ —á–∏—Å–µ–ª $(a, b)$ ‚Äì —Ü–µ –∑–∞—à–∏—Ñ—Ä–æ–≤–∞–Ω–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è.

–£ –Ω–∞—à–æ–º—É –≤–∏–ø–∞–¥–∫—É, –∫ –ù–ï –ú–ê–Ñ –¥—ñ–ª–∏—Ç–∏—Å—è –Ω–∞ 2,3,17
–Ø —É —Å–≤–æ—î–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ –±—Ä–∞–≤ —á–∏—Å–ª–∞ 49, 65, 101
–û–±—á–∏—Å–ª–µ–Ω–Ω—è –ê —ñ –ë –≤ –∫–æ–¥—ñ –Ω–∏–∂—á–µ
–í –ö–û–î–Ü –ó–ê–ö–û–ú–ï–ù–¢–û–í–ê–ù–û –†–Ø–î–ö–ò –ó –í–í–û–î–û–ú –ö –í–Ü–î –ö–û–†–ò–°–¢–£–í–ê–ß–ê, –º–æ–∂–Ω–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ —ñ —ó—Ö

```python
import math

# –±—ã—Å—Ç—Ä–æ–µ –≤–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å –ø–æ –º–æ–¥—É–ª—é
def modpow(base, exp, mod):
    result = 1
    base %= mod
    while exp > 0:
        if exp % 2 == 1:
            result = (result * base) % mod
        base = (base * base) % mod
        exp //= 2
    return result

# –î–∞–Ω–Ω—ã–µ 
p = 103
g = 5
y = 44
m = 3

k = int(input("–í–≤–µ–¥–∏—Ç–µ k (1 < k < 102): ")) 
# k = 101

if not (1 < k < p - 1):
    print("k –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ 2..101")
    exit()

if math.gcd(k, p - 1) != 1:
    print("k –ù–ï –ø–æ–¥—Ö–æ–¥–∏—Ç (gcd(k, p-1) != 1)")
    exit()
else:
    print("k –ø–æ–¥—Ö–æ–¥–∏—Ç!")

# –®–∏—Ñ—Ä–æ–≤–∞–Ω–∏–µ
a = modpow(g, k, p)
b = (modpow(y, k, p) * m) % p

print("a = g^k mod p =", a)
print("b = y^k * m mod p =", b)
print("–ó–∞—à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ (a, b) =", (a, b))
```

–ö–æ–¥ –∑—Ä–æ–±–ª–µ–Ω–∏–π, —â–æ –≤—Å–µ —Ç—Ä–µ–±–∞ –≤–≤–æ–¥–∏—Ç–∏ —Å–∞–º–æ—Å—Ç—ñ–π–Ω–æ, –ë—É–¥–µ –∑–∞–∫–æ–º–µ–Ω—Ç–æ–≤–∞–Ω–∞ –º–æ–∂–ª–∏–≤—ñ—Å—Ç—å –∑–∞ –≤–∂–µ –∑–∞–¥–∞–Ω–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏

### –ó–∞–≤–¥–∞–Ω–Ω—è 2: –∑–∞–¥–∞—á–∞ –∑ –ø—Ä–æ–≥—Ä–∞–º—É–≤–∞–Ω–Ω—è

–û–±–µ—Ä—ñ—Ç—å —Ç–∞ —Ä–æ–∑–≤'—è–∂—ñ—Ç—å **–æ–¥–Ω—É** –∑ –∑–∞–¥–∞—á –¥–ª—è –ø—Ä–æ–≥—Ä–∞–º—É–≤–∞–Ω–Ω—è 8.11, 8.12, 9.10, 13.12, 14.3, 14.14
–∑ –∞—Ä–∫—É—à—ñ–≤ –¥–ª—è –ø—Ä–∞–∫—Ç–∏—á–Ω–∏—Ö –∑–∞–Ω—è—Ç—å (–¥–∏–≤. —É Google Classroom)

9.10. –ù–∞–ø–∏—ài—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ä–æ–∑—à–∏—Ä–µ–Ω–æ–≥–æ –∞–ª–≥–æ-
—Ä–∏—Ç–º—É –ï–≤–∫–ªi–¥–∞:

1) —Ä–æ–∑–≤‚Äô—è–∑—É—î –ªi–Ωi–π–Ωi –∫–æ–Ω–≥—Ä—É–µ–Ω—Üi—ó –≤–∏–≥–ª—è–¥—É ax ‚â° b mod n;
2) –∑–Ω–∞—Ö–æ–¥–∏—Ç—å –æ–±–µ—Ä–Ω–µ–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –¥–æ –¥–∞–Ω–æ–≥–æ —É –ø–æ–ªi –∫–ª–∞—Åi–≤ –ª–∏—à–∫i–≤ Zp.

–í–≤—ñ–¥ —É —Ç–µ—Å—Ç–æ–≤–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ: 12 21 3 3 7

```python
def extended_gcd(a: int, b: int):
    if b == 0:
        return a, 1, 0
    g, x1, y1 = extended_gcd(b, a % b)
    x = y1
    y = x1 - (a // b) * y1
    return g, x, y


def solve_congruence(a: int, b: int, n: int):
    a %= n
    b %= n

    g, x0, y0 = extended_gcd(a, n)

    if b % g != 0:
        return None

    a1 = a // g
    b1 = b // g
    n1 = n // g

    g2, inv_a1, _ = extended_gcd(a1, n1)

    x_base = (inv_a1 * b1) % n1

    solutions = []
    for k in range(g):
        xk = (x_base + k * n1) % n
        solutions.append(xk)

    return sorted(solutions)


def inverse_mod(a: int, p: int):
    a %= p
    g, x, y = extended_gcd(a, p)
    if g != 1:
        return None
    return x % p


# —Ä–µ—à–µ–Ω–∏–µ –ª–∏–Ω–µ–π–Ω–æ–π –∫–æ–Ω–≥—Ä—É–µ–Ω—Ü–∏–∏
a = int(input("a = "))
b = int(input("b = "))
n = int(input("n = "))

solutions = solve_congruence(a, b, n)
print("–†–æ–∑–≤'—è–∑–∫–∏ –ª—ñ–Ω—ñ–π–Ω–æ—ó –∫–æ–Ω–≥—Ä—É–µ–Ω—Ü—ñ—ó:")
print(solutions if solutions is not None else "–ù–µ–º–∞—î —Ä–æ–∑–≤'—è–∑–∫—ñ–≤")

#–æ–±—Ä–∞—Ç–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤ Z_p
a2 = int(input("\na (–¥–ª—è –æ–±–µ—Ä–Ω–µ–Ω–æ–≥–æ –≤ Z_p) = "))
p = int(input("p = "))

inv = inverse_mod(a2, p)
print("–û–±–µ—Ä–Ω–µ–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç:")
print(inv if inv is not None else "–û–±–µ—Ä–Ω–µ–Ω–æ–≥–æ –Ω–µ —ñ—Å–Ω—É—î")
```

# 1.–ß–æ–º—É —Å—Ö–µ–º–∞ –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏ –ø—Ä–∞—Ü—é—î

–£ —Å—Ö–µ–º—ñ RSA –∫–ª—é—á—ñ –∑–∞–¥–æ–≤–æ–ª—å–Ω—è—é—Ç—å —Å–ø—ñ–≤–≤—ñ–¥–Ω–æ—à–µ–Ω–Ω—è:
$ed \equiv 1\ {\rm mod}\ \varphi(n)$

–¥–µ $e$ ‚Äî –≤—ñ–¥–∫—Ä–∏—Ç–∞ –µ–∫—Å–ø–æ–Ω–µ–Ω—Ç–∞, $d$ ‚Äî –∑–∞–∫—Ä–∏—Ç–∞ –µ–∫—Å–ø–æ–Ω–µ–Ω—Ç–∞, $n = pq$.

–ü—ñ–¥–ø–∏—Å –ë–æ–±–∞:
$s = m^d\ {\rm mod}\ n$

–ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –ø—ñ–¥–ø–∏—Å—É:
$s^e\ {\rm mod}\ n = (m^d)^e\ {\rm mod}\ n = m^{de}\ {\rm mod}\ n$

–û—Å–∫—ñ–ª—å–∫–∏ $ed \equiv 1\ {\rm mod}\ \varphi(n)$, —Ç–æ:
$m^{de} \equiv m\ {\rm mod}\ n$

–ó–≤—ñ–¥—Å–∏:
‚Äî —Ç—ñ–ª—å–∫–∏ —Ç–æ–π, —Ö—Ç–æ –∑–Ω–∞—î $d$, –º–æ–∂–µ —Å—Ç–≤–æ—Ä–∏—Ç–∏ –ø—ñ–¥–ø–∏—Å;  
‚Äî –±—É–¥—å-—Ö—Ç–æ, —Ö—Ç–æ –º–∞—î –≤—ñ–¥–∫—Ä–∏—Ç–∏–π –∫–ª—é—á $(n, e)$, –º–æ–∂–µ –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏ –ø—ñ–¥–ø–∏—Å.


–ö–û–î
‚Äî –≥–µ–Ω–µ—Ä–∞—Ü—ñ—è –∫–ª—é—á—ñ–≤ RSA;  
‚Äî –ø—ñ–¥–ø–∏—Å –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è: $s = m^d\ {\rm mod}\ n$;  
‚Äî –ø–µ—Ä–µ–≤—ñ—Ä–∫–∞ –ø—ñ–¥–ø–∏—Å—É: $s^e\ {\rm mod}\ n = m$.

```python
from sympy import randprime, mod_inverse
import random

# --- –≥–µ–Ω–µ—Ä—É–≤–∞–Ω–Ω—è –∫–ª—é—á—ñ–≤ RSA ---
def generate_keys(bits=16):
    # –≥–µ–Ω–µ—Ä—É—î–º–æ –¥–≤–∞ –≤–∏–ø–∞–¥–∫–æ–≤–∏—Ö –ø—Ä–æ—Å—Ç–∏—Ö
    p = randprime(2**(bits-1), 2**bits)
    q = randprime(2**(bits-1), 2**bits)

    n = p * q
    phi = (p - 1) * (q - 1)

    # –≤–∏–±–∏—Ä–∞—î–º–æ e –≤–∑–∞—î–º–Ω–æ –ø—Ä–æ—Å—Ç–µ –¥–æ phi
    while True:
        e = random.randrange(2, phi)
        if gcd(e, phi) == 1:
            break

    d = mod_inverse(e, phi)
    return (n, e), (p, q, d)

# --- –∞–ª–≥–æ—Ä–∏—Ç–º –ï–≤–∫–ª—ñ–¥–∞ ---
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# --- –ø—ñ–¥–ø–∏—Å ---
def sign(message, private_key):
    p, q, d = private_key
    n = p * q
    return pow(message, d, n)

# --- –ø–µ—Ä–µ–≤—ñ—Ä–∫–∞ –ø—ñ–¥–ø–∏—Å—É ---
def verify(message, signature, public_key):
    n, e = public_key
    return pow(signature, e, n) == message

# --- –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü—ñ—è ---
public_Bob, private_Bob = generate_keys()
print("–í—ñ–¥–∫—Ä–∏—Ç–∏–π –∫–ª—é—á –ë–æ–±–∞:", public_Bob)
print("–ó–∞–∫—Ä–∏—Ç–∏–π –∫–ª—é—á –ë–æ–±–∞:", private_Bob)

m = 1234  # –¥–µ—è–∫–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è
s = sign(m, private_Bob)
print("–ü—ñ–¥–ø–∏—Å:", s)

print("–ü–µ—Ä–µ–≤—ñ—Ä–∫–∞:", verify(m, s+2, public_Bob))
```
